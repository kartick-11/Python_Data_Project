```python
#Importing Libraries
import ast
import pandas as pd
from datasets import load_dataset
import matplotlib.pyplot as plt

#Loading the dataset
dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()

#Data Cleanup
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
df['job_skills'] = df['job_skills'].apply(lambda x: ast.literal_eval(x) if pd.notna(x) else x)
```

```python
df_DA = df[df['job_title_short'] == 'Data Analyst'].copy()

df_DA
```

```python
df_DA['job_posted_month_no'] = df_DA['job_posted_date'].dt.month

df_DA_exploded = df_DA.explode('job_skills')

df_da_pivot = df_DA_exploded.pivot_table(index='job_posted_month_no', columns='job_skills', aggfunc='size', fill_value=0)

df_da_pivot
```

```python
df_DA['job_posted_month_no'] = df_DA['job_posted_date'].dt.month

df_DA_exploded = df_DA.explode('job_skills')

df_da_pivot = df_DA_exploded.pivot_table(index='job_posted_month_no', columns='job_skills', aggfunc='size', fill_value=0)

df_da_pivot.plot(kind='line')
plt.legend().set_visible(False)
```

```python
df_da_pivot.loc['total'] = df_da_pivot.sum()

df_da_pivot = df_da_pivot[df_da_pivot.loc['total'].sort_values(ascending=False).index]

df_da_pivot = df_da_pivot.drop('total')

df_da_pivot

```

```python
df_da_pivot.iloc[:, :5]
```

```python
df_da_pivot.iloc[:, :5].plot(kind='line')
plt.title('Top 5 Skills for Data Analyst Jobs Over Time')
plt.xlabel('Month')
plt.ylabel('Number of Job Postings')
plt.legend(title='Skills', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```
