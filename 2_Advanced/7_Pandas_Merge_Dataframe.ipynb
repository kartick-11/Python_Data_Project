```python
#Importing Libraries
import pandas as pd
from datasets import load_dataset
import matplotlib.pyplot as plt

#Loading the dataset
dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()

#Data Cleanup
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
```

```python
df_US = df[df['job_country'] == 'United States'].copy()
df_US['job_posted_month'] = df_US['job_posted_date'].dt.strftime('%B')
df_US_pivot = df_US.pivot_table(index='job_posted_month', columns='job_title_short', aggfunc='size')
df_US_pivot = df_US_pivot.reset_index()
df_US_pivot['month_no'] = pd.to_datetime(df_US_pivot['job_posted_month'], format='%B').dt.month
df_US_pivot = df_US_pivot.sort_values('month_no')
df_US_pivot = df_US_pivot.set_index('job_posted_month')
df_US_pivot = df_US_pivot.drop(columns='month_no')

df_US_pivot
```

```python
df_US_software_pivot = pd.read_csv("https://lukeb.co/software_csv", index_col = 'job_posted_month')

df_US_software_pivot
```

```python
df_US_mergerd = df_US_pivot.merge(df_US_software_pivot, on='job_posted_month')
```

```python
top_5 = (df_US_mergerd
        .sum()
        .sort_values(ascending=False)
        .head()
        .index.tolist())

top_5
```

```python
df_US_mergerd[top_5].plot(kind='line')
plt.title('Top 5 Job Titles Posted in the US by Month')
plt.xlabel('2023 - Months')
plt.ylabel('Number of Job Postings')
plt.ylim(0, 25000)
plt.legend(title='Job Titles')
plt.show()
```
